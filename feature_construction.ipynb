i want to create a feature that there two columns first is about sibling/childs and secound is about parents so by  adding this i make a new coulumn family size ---------> sibsp+parch and then make it as a family type

```python
new_df['family_size'] = df['SibSp']+df['Parch']
```

```python
new_df.head()
```

```python
new_df.sample(10)
```

```python
def myfunc(size):
  if size ==1:
    return 1 #alone
  elif 1<size<=4:
    return 2 #small
  else:
    return 3 #big
```

```python
myfunc(5)
```

```python
new_df['family_type'] = new_df['family_size'].apply(myfunc)
```
