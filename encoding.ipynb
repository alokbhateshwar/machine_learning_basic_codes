```python
educational_level = ['School','HighSchool','Gradute']
```

```python
import random
df['Education']= [random.choice(educational_level) for _ in range(len(df))]
```

```python
df.sample(5)
```

```python
from sklearn.preprocessing import OrdinalEncoder
model_oe= OrdinalEncoder(categories=[['School','HighSchool','Gradute']])
```

```python
model_oe.fit(df[['Education']])
educational_encoded=model_oe.transform(df[['Education']])
```

```python
educational_encoded
```

```python
df['Education'] = educational_encoded
```
