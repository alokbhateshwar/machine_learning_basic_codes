```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
print(data.DESCR)
```

```python
x=data.data
y=data.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#first we do with linearRegression for the comparision 
from sklearn.linear_model import  LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
prediciton1= lr.predict(x_test)
print(prediciton1)
```

```python
from sklearn.metrics import  r2_score,mean_squared_error
print("R2_Score is :",r2_score(y_test,prediciton1))
print("Root Mean Squre Error:",np.sqrt(mean_squared_error(y_test,prediciton1)))
```

```python
#now we done this with Ridge 
from sklearn.linear_model import  Ridge
r=Ridge(alpha=0.001)
r.fit(x_train,y_train)
```

```python
prediciton2 = r.predict(x_test)
```

```python
print(prediciton2)
```

```python
from sklearn.metrics import  r2_score,mean_squared_error
print("R2_Score is :",r2_score(y_test,prediciton2))
print("Root Mean Squre Error:",np.sqrt(mean_squared_error(y_test,prediciton2)))
```

```python
#dont use it because of overfiting 
best_r2 =-1.0
for i in range(1000):
    ri = Ridge(alpha=i)
    ri.fit(x_train,y_train)
    y_pred= ri.predict(x_test)
    R2=r2_score(y_test,y_pred)
if R2 > best_r2:
    best_r2 = R2
    alpha = i
    print("the best value of alpha:",alpha)
```
