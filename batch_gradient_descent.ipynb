```python
import numpy as np
class GDRegression:
  def __init__(self,learning_rate=0.1,epochs=100):
    self.epochs=epochs
    self.lr=learning_rate
    self.intercept_=None
    self.coef_=None
  def fit(self,x_train,y_train):
    #random coef_
    self.intercept_=0 #here intercept is a single number so we can initialize with it a single number
    # we have multiple coef_ so we cant initialize with a single numer so we use here numpy arrays which size depends on the size of x_train
    self.coef_=np.ones(x_train.shape[1])
    print("intercepts B0:",self.intercept_)
    print("coffecint B1,B2,B3........B10:",self.coef_)

    for i in range(self.epochs):
      #now we update all the coef and intercepts
      y_hat = np.dot(x_train,self.coef_)+self.intercept_
      intercept_der = -2*np.mean(y_train-y_hat)
      self.intercept_= self.intercept_-(self.lr*intercept_der)

      coef_der = -2*np.dot((y_train-y_hat),x_train)/x_train.shape[0]
      self.coef_=self.coef_-(self.lr*coef_der)
    print("intercept B0:",self.intercept_)
    print("coef B1.....B10:",self.coef_)


  def predict(self,x_test):
     return np.dot(x_test,self.coef_)+self.intercept_
```

```python
gdr= GDRegression(epochs=11000,learning_rate=0.2)
```

```python
gdr.fit(x_train,y_train)
```

```python
y_pred =gdr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
