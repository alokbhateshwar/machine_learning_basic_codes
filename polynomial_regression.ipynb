```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
x=6*np.random.rand(200,1)-3
y=0.8 *x**2+0.9*x+2+np.random.randn(200,1)
```

```python
plt.plot(x,y,'b.')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print(y_pred)
```

```python
print("coeffecient",lr.coef_)
print("intercepts",lr.intercept_)
```

```python
r2=r2_score(y_test,y_pred)
print("acrurecy is",r2*100,"%")
```

```python
plt.plot(x,y,"r.")
plt.plot(x_train,lr.predict(x_train),"b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

so we can see that this is not predict accurately so we transform into polynomial features which findes the values of x^0,x^1,x^2..... and all and imp thing is we can apply features on only train data x_train,x_test

```python
poly =PolynomialFeatures(degree=2)
```

```python
x_train_trans= poly.fit_transform(x_train)
x_test_trans =poly.fit_transform(x_test)
```

```python
lr.fit(x_train_trans,y_train)
```

```python
lr.predict(x_test_trans)
```

```python
plt.plot(x,y,"r.")
plt.plot(x_train,lr.predict(x_train_trans),"b.")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
